All variables are of object data type and thus strings. For better use there is need to convert them to numerical to be used later in complex calculations and models.

```python
df.describe()
```

```python
# Check for missing values
print(df.isnull().sum()) # 2 missing values on rating_count
df[df.rating_count.isnull()] # print rows with missing fields
```

```python
# Remove rows with missing values in the rating_count column
# They are only 2, they wont affect the datase
df.dropna(subset=['rating_count'], inplace=True)
print(df.isnull().sum())
```

```python
#format numerial columns, remove symbols from rows
df['discounted_price'] = df['discounted_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
df['actual_price'] = df['actual_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%','').astype(float)/100
```

```python
df.rating_count = df.rating_count.astype(float)
```

There is a row with '24,269' on rating_count.

```python
#df['actual_price'] = df['actual_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
print(df[[df.rating_count]== '24,269'])
```

```python
# Function to check for special characters
def has_special_chars(s):
    return any(char.isdigit() or char.isalpha() or char.isspace() for char in s)

# Filter rows with special characters
special_rows = df.rating_count.map(lambda x: not has_special_chars(x)).any()

# Print rows with special characters
print(df[special_rows])
```
